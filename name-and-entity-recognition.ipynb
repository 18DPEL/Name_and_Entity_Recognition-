# As we know that our input data is in text format so we have to convert them into csv format

```python
cols = ['labels','words']
plaintext_files = pd.DataFrame()
address = glob.glob('/kaggle/input/data-bae-dataset/Training Data for Problem 2.txt')
count = 0
for i in address:
    with open(i, "rb") as data_of_files:
        data = pd.read_csv(data_of_files, sep='\t', header=None, names=cols)
        plaintext_files = pd.concat([plaintext_files, data[['labels', 'words']]], axis=0)
        count = count + 1
        if count == 493:
            break

data= plaintext_files.rename(columns={0:'labels', 1:'words'})
data.head(50)
```

# text preprocessing

here i am checking is there null value available or not inside to the data

```python
data.isnull().sum()
```

droping the row who have null value

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

here i am defining the independent and dependent variable 

```python
x=data[['words']]
y=data['labels']
```

```python
x
```

Spliting the data into train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(x_train)
```

```python
train_data=pd.DataFrame({"words":x_train["words"],"labels":y_train})
test_data=pd.DataFrame({"words":x_test["words"],"labels":y_test})
```

```python
train_data
```

# Model_training

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label=data["labels"].unique().tolist()
label
```

# Here i am defining the some hyperparameter 

```python
args=NERArgs()
args.num_train_epochs=2
args.learning_rate=0.0001
args.overwrite_output_dir=True
args.train_batch_size=32
args.eval_batch_size=32
```

```python
train_data["sentence_id"] = range(len(train_data))
test_data["sentence_id"] = range(len(test_data))
```

```python
model=NERModel('bert','bert-base-cased',labels=label,args=args)
```

```python
model=NERModel('bert','bert-base-cased',labels=label,args=args)
model.train_model(train_data,eval_data=test_data,acc=accuracy_score)
```

As you can see here after doing some hyper parameter tuning  , i reduce the loss arround **0.21** and accuracy of the model is **0.93%**

here i am saving the model

```python
result,model_outputs,preds_list=model.eval_model(test_data)
```
